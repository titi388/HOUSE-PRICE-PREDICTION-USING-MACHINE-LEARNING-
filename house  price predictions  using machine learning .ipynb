```python
# convert the garageYrBlt to object data type
train_df_d["GarageYrBlt"] = train_df_d["GarageYrBlt"].astype(object)

mode= train_df_d["GarageYrBlt"].mode([0])
train_df_d["GarageYrBlt"].fillna(mode, inplace=True)
```

```python
for i in train_df_d[c.index].select_dtypes(include=[int,float]).columns:
    mode = train_df_d[i].mode()[0]
    train_df_d[i].fillna(mode, inplace=True)
```

```python
# separate the target from the features
x = train_df_d.copy().drop(columns= ["Id"])
y= x.pop("SalePrice")
```

```python
x.select_dtypes(include=[object])
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
ordinal_variable=  x[["LotShape", "LandContour", "LandSlope", "OverallQual", "OverallCond", "ExterQual", "HouseStyle",
"ExterQual", "ExterCond", "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2",
"HeatingQC", "KitchenQual", "GarageQual", "GarageCond", 'SaleCondition']]
le= LabelEncoder()
for i in nominal_variable:
    x[i] = le.fit_transform(x[i])
```

# enconding it one after the other
le = LabelEncoder()
x["MSZoning"] = le.fit_transform(x["MSZoning"])

```python
x.select_dtypes(include=[object])
```

```python
# convert the garageYrBlt to float data type
x["GarageYrBlt"] = x["GarageYrBlt"].astype(float)
```

```python
#  remaining category is nominal variable 
nominal_variable=x.select_dtypes(include=[object]).columns
x_encode=pd.get_dummies(x, columns=nominal_variable)
nominal_variable
```

```python
# split the data into train and validation set
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_encode, y, random_state=42,test_size=0.2)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb= GradientBoostingRegressor()
model_ = gb.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
val_predict = model_.predict(x_val)
eval_r_1 = r2_score(y_val, val_predict)
eval_m_1 = mean_absolute_error(y_val, val_predict)
print(f"r2_score value for{model_} is {eval_r_1}")
print(f"mean_absolute_error value for {model_} is  {eval_m_1}")
```
